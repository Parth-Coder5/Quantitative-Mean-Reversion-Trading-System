```python
stocks = [
    'KOTAKBANK.NS',
    'AXISBANK.NS',
    'CANBK.NS',
    'HDFCBANK.NS',
    'ICICIBANK.NS',
    'SBIN.NS',
    'IDBI.NS',
    'AUBANK.NS',
    'IDFCFIRSTB.NS',
    'CUB.NS'
]
```

```python
import backtest.engine as eng
import strategy.indicators as ind
import strategy.signals as sig

importlib.reload(eng)
importlib.reload(ind)
importlib.reload(sig)

run_backtest = eng.run_backtest
add_indicators = ind.add_indicators
calculate_rsi = ind.calculate_rsi
signal_generator = sig.signal_generator

# Rebuild a clean DataFrame so notebook re-runs don't break on stale state
for stock in stocks:

    print(f"\n========== {stock} ==========")

    try:
        df = fetch_data(stock, '2021-01-01', '2026-01-01')
    except Exception as e:
        print(f"Skipping {stock}: data fetch failed ({e})")
        continue

    if df is None or df.empty:
        print(f"Skipping {stock}: no data returned")
        continue

    df = add_indicators(df)
    df['RSI'] = calculate_rsi(df['Close'])
    df = signal_generator(df)
    run_backtest(df)
```
